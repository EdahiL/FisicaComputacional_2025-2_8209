```python
def f1(x):
    return x

def f2(x):
    return 2 - np.exp(-x) 
```

```python
X = np.linspace(-1.5,2,100)
Y1 = f1(X)
Y2 = f2(X)
plt.plot(X,Y1)
plt.plot(X,Y2)
plt.grid()
```

```python
def g1(x):
    return x

def g2(x):
    return np.tan(x) 

X = np.linspace(-0.00005,.00005,1000)
Y1 = g1(X)
Y2 = g2(X)
plt.plot(X,Y1,'b:')
# plt.plot(X,Y2,'r.')
plt.plot(X,Y2)
plt.grid()
```
