```python
# # sp.integrate.trapezoid?
# def gaussxwab(N,a,b):
#     x,w = sp.special.roots_legendre(N)
#     return 0.5*(b-a)*x+0.5*(b+a),0.5*(b-a)*w

def Integral(f,a,b,N):
    x,w = sp.special.roots_legendre(N)
    xp = 0.5*(b-a)*x+0.5*(b+a)
    wp = 0.5*(b-a)*w

    S = 0
    for k in range(N):
        S = S + wp[k]*f(xp[k])
    # S = sum(f(xkp)*wkp)

    return S
```

```python
N=50
a=-10
b=10
xk,wk = sp.special.roots_legendre(N)
```

```python
xkp = 0.5*(b-a)*xk+0.5*(b+a)
wkp = 0.5*(b-a)*wk
```
