```python
def Derivada1a(f,x0,h):
    df = (f(x0+h/2) - f(x0-h/2))/h
    return df

def Derivada2a(f,x0,h):
    df = (f(x0+h) -2*f(x0) + f(x0-h))/h**2
    return df

def ParcialX(f,x0,y0,h):
    dfx = (f(x0+h/2,y0) - f(x0-h/2,y0))/h
    return dfx

def ParcialY(f,x0,y0,h):
    dfy = (f(x0,y0+h/2) - f(x0,y0-h/2))/h
    return dfy
```
